```python
import numpy as np
import pandas as pd
```

```python
# Wczytywanie danych
train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")
```

```python
# Dodawanie roku, miesiÄ…ca i dnia

# Train
train_year = train.timestamp.str[:4]
train_month = train.timestamp.str[5:7].astype(float)
train_day = train.timestamp.str[8:10].astype(float)

train['year'] = train_year
train['month'] = train_month
train['day'] = train_day

# Test
test_year = test.timestamp.str[:4]
test_month = test.timestamp.str[5:7].astype(float)
test_day = test.timestamp.str[8:10].astype(float)

test['year'] = test_year
test['month'] = test_month
test['day'] = test_day
```

```python
# Save
train.to_csv("../mdata/data_train.csv")
test.to_csv("../mdata/data_test.csv")
```
