```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Wczytywanie danych
train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")
```

```python
ntrain = train.isnull().any()
nan_vals_train = ntrain[ntrain == 1].index.tolist()
```

```python
ntest = test.isnull().any()
nan_vals_test = ntest[ntest == 1].index.tolist()
```

```python
set(nan_vals_test).difference(set(nan_vals_train))
```
